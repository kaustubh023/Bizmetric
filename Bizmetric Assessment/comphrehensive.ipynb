```python
# 6 Extract all vowels from a given string using list comprehension. 
text = "Kaustubh Pawar"

vowels = [char for char in text if char.lower() in "a e i o u"]

print(vowels)
```

```python
# 7.	Flatten a 2D list using list comprehension. 

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

flatlist = [item for row in matrix for item in row]

print(flatlist)
```

```python
# 8.	Replace all negative numbers in a list with 0 using list comprehension. 

numbers = [-5, 3, -2, 7, -8, 10]

list1 = [num if num >= 0 else 0 for num in numbers]

print(list1)
```

```python
# 9.	Given a list of words, create a list of lengths of each word. 

words = ["kaustubh", "peeyush", "gaurav", "pranav","dhanesh"]

lengths = [len(word) for word in words]

print(lengths)
```

```python
# 10.	Filter out words that start with the letter 'A' or 'a'. 
words = ["kaustubh", "peeyush", "gaurav", "pranav","dhanesh"]
names_a = [word for word in words if not word.lower().startswith('a')]

print(names_a)
```

```python
# 11.	From a list of numbers, generate a list of “even” or “odd” strings using list comprehension. 

numbers = [1, 2, 3, 4, 5, 6]
result = ["even" if num % 2 == 0 else "odd" for num in numbers]

print(result)
```

```python
# 12.	Create a list of numbers divisible by both 3 and 5 in range 1–100. 

number = [num for num in range(1, 101) if num % 3 == 0 and num % 5 == 0]

print(number)
```

```python
# 13.	Write a nested list comprehension to generate a multiplication table for 1–5. 

table = [[i * j for j in range(1, 10)] for i in range(1, 6)]

print(table)
```

```python
# 14. Convert a dictionary’s keys into a list using list comprehension. 
student = {
    "name": "Kaustubh",
    "age": 22,
    "course": "Python"
}
keys_list = [key for key in student]

print(keys_list)
```

```python
# 15.	Extract numeric digits from a string using list comprehension. 
text = "kwehuh23u9ruhwih82u939"
digits = [char for char in text if char.isdigit()]

print(digits)
```

```python
# 16.	Use list comprehension to remove all spaces from a string. 

text = "Kaustubh Pawar Python Developer"
remove_spaces = ''.join([char for char in text if char != ' '])

print(remove_spaces)
```

```python
# 17.	Create a list of characters that appear more than once in a string. 
text = "programming"

chars = list({char for char in text if text.count(char) > 1})

print(chars)
```

```python
# 18.	From a list of sentences, generate a list of all words (split using list comprehension). 

sentences = [
    "Python is powerful",
    "SQL is used for databases"
]
words = [word for sentence in sentences for word in sentence.split()]

print(words)
```

```python
# 19.	Create a list of unique elements from a list using list comprehension + condition. 

numbers = [1, 2, 2, 3, 4, 4, 5, 1, 6]
unique=[]
unique = [x for x in numbers if not (x in unique or unique.append(x))]

print(unique)
```

```python
# 20.	Generate all pairs (x, y) where x is from list A and y is from list B (cartesian product).

A = [1, 2, 3]
B = ['a', 'b']

pairs = [(x, y) for x in A for y in B]

print(pairs) 
```
