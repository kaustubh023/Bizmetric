```python
# Textbook prices
textbooks = {
    "Hindi":    {"1-4": 60,  "5-8": 100, "9-10": 150},
    "Marathi": {"1-4": 60,  "5-8": 100, "9-10": 150},
    "English": {"1-4": 80,  "5-8": 100, "9-10": 150},
    "Science": {"1-4": 90,  "5-8": 120, "9-10": 200},
    "Maths":   {"1-4": 100, "5-8": 140, "9-10": 250}
}

# Notebook prices
notebooks = {
    "Square":       {100: 40, 200: 70},
    "4 Lines":      {100: 30, 200: 50},
    "2 Lines":      {100: 30, 200: 50},
    "Single Line":  {100: 60, 200: 100},
    "A4":           {100: 100, 200: 180}
}
```

```python
bill = []
total_amount = 0

print(" Welcome to Stationery Shop ")

while True:
    print("\nWhat do you want to buy?")
    print("1. Textbook")
    print("2. Notebook")
    print("3. Finish & Generate Bill")

    choice = input("Enter choice (1/2/3): ")

    # ---------------- TEXTBOOK ----------------
    if choice == '1':
        standard = int(input("Enter Standard (1-10): "))

        if 1 <= standard <= 4:
            group = "1-4"
        elif 5 <= standard <= 8:
            group = "5-8"
        elif 9 <= standard <= 10:
            group = "9-10"
        else:
            print("Invalid standard")
            continue

        print("Available Textbooks:")
        for book in textbooks:
            print(book)

        book_name = input("Enter book name: ")
        qty = int(input("Enter quantity: "))

        price = textbooks[book_name][group]
        cost = price * qty

        bill.append((book_name, qty, price, cost))
        total_amount += cost

        print(f"Added {book_name} × {qty} → ₹{cost}")

    # ---------------- NOTEBOOK ----------------
    elif choice == '2':
        print("Available Notebooks:")
        for nb in notebooks:
            print(nb)

        nb_type = input("Enter notebook type: ")
        pages = int(input("Enter pages (100 or 200): "))
        qty = int(input("Enter quantity: "))

        price = notebooks[nb_type][pages]
        cost = price * qty

        bill.append((nb_type, qty, price, cost))
        total_amount += cost

        print(f"Added {nb_type} Notebook × {qty} → ₹{cost}")

    # ---------------- FINISH ----------------
    elif choice == '3':
        print("\n BILL SUMMARY")
        print("-" * 30)

        for item in bill:
            print(f"{item[0]} × {item[1]} @ ₹{item[2]} = ₹{item[3]}")

        print("-" * 30)
        print("Total Amount:", total_amount)

        confirm = input("Confirm purchase? (yes/no): ").lower()

        if confirm == 'yes':
            print(" Bill Generated Successfully!")
            print("Thank you for shopping ")
            break
        else:
            print(" Bill Cancelled")
            break

    else:
        print("Invalid choice")
```
