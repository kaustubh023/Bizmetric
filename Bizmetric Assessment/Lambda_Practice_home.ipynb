```python
# 16.	Use lambda to sort a list of dictionaries by a key (like "age"). 

people = [
    {"name": "Kaustubh", "age": 22},
    {"name": "Rahul", "age": 25},
    {"name": "Anita", "age": 20}
]
sort = sorted(people, key=lambda person: person["age"])

print(sort)
```

```python
# 20.	Write a lambda to remove duplicates from a list using filter + set. 

numbers = [1, 2, 2, 3, 4, 4, 5, 1, 6]

s= set()
unique = list(filter(lambda x: not (x in s or s.add(x)), numbers))

print(unique)
```

```python
# 26.	Use lambda with map to convert Celsius to Fahrenheit. 

celsius = [0, 20, 30, 37, 100]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print(fahrenheit)
```
